## Imports

```python
import pandas as pd
import folium
from IPython.display import display
import io
from dataclasses import dataclass
import random
import numpy as np
from copy import copy
```

## Utility functions

```python
def draw(df, tour=None):
    """
    Draws a map with location markers and lines connecting points according
    to a specified tour.

    Parameters:
    -----------
    df: pd.DataFrame
        DataFrame containing at minimum 'Latitude' and 'Longitude' columns.
    tour: list
        List of indices specifying the order to connect points.
        If None, no lines are drawn.
    """
    center_lat = 52.1326
    center_lon = 5.2913
    zoom_start = 7
    width = 400
    height = 420
    point_radius = 0.1
    point_color = "red"
    line_color = "blue"
    line_weight = 2

    fmap = folium.Map(
        location=[center_lat, center_lon],
        zoom_start=zoom_start,
        tiles="Cartodb positron",
        width=width,
        height=height,
    )

    for _, row in df.iterrows():
        popup = f"{row['Name']}" if "Name" in df.columns else None
        folium.CircleMarker(
            location=[row["Latitude"], row["Longitude"]],
            radius=point_radius,
            popup=popup,
            color=point_color,
            fill=True,
            fill_color=point_color,
            fill_opacity=0.6,
        ).add_to(fmap)

    if tour is not None:
        coordinates = []
        for idx in tour:
            row = df.iloc[idx]
            coordinates.append([row["Latitude"], row["Longitude"]])

        if tour[0] != tour[-1]:
            coordinates.append(coordinates[0])

        folium.PolyLine(
            locations=coordinates,
            color=line_color,
            weight=line_weight,
            opacity=0.7,
        ).add_to(fmap)

    return fmap
```

```python
df = pd.read_csv(io.StringIO(DATA))

# Example usage:
tour = [0, 2, 1, 3]  # Example: visit in this order, returning to start
draw(df, tour=tour)
```

### Haversine distance

```python
def haversine(coords: list[tuple[int, int]]) -> np.ndarray:
    """
    Converts coordinates to a Haversine distance matrix (in meters).
    """

    from sklearn.metrics.pairwise import haversine_distances
    from math import radians

    RADIUS_EARTH = 6371000 # meters

    coords_rad = np.array(
        [[radians(lat), radians(lon)] for lat, lon in coords]
    )
    dist_matrix = haversine_distances(coords_rad)
    dist_meters = (dist_matrix * RADIUS_EARTH).round() # rads to meters
    return dist_meters
```

### ProblemData

```python
@dataclass
class ProblemData:
    distances: list[list[int]]

    @property
    def num_clients(self):
        return len(self.distances)
```

### Cost function

```python
def cost(tour, data):
    """
    Computes the cost of a tour.
    """
    nodes = tour + [tour[0]]
    frm = nodes[:-1]
    to = nodes[1:]
    return data.distances[frm, to].sum()
```

## Nearest neighbor

```python
def nearest_neighbor(data: ProblemData) -> list[int]:
    current = 0
    tour = [current]
    unvisited = set(range(data.num_clients)) - {current}

    while unvisited:
        neighbor = min(
            unvisited, key=lambda client: data.distances[current][client]
        )
        unvisited -= {neighbor}
        tour.append(neighbor)
        current = neighbor

    return tour
```

```python
coords = df[["Latitude", "Longitude"]].values
data = ProblemData(haversine(coords))
```

```python
tour = nearest_neighbor(data)
cost(tour, data)
```

```python
fmap = draw(df, tour=tour)
display(fmap)
```

## Large neighborhood search

```python
def large_neighborhood_search(
    data: ProblemData,
    num_iterations: int,
    num_removals: int = 5,
) -> list[int]:
    curr_tour = nearest_neighbor(data)
    curr_cost = cost(curr_tour, data)

    for _ in range(num_iterations):
        new_tour = copy(curr_tour)

        # Destroy
        start_idx = random.randint(0, data.num_clients - num_removals)
        removed = new_tour[start_idx:start_idx+num_removals] # [4, 5, 6, 7, 8]
        del new_tour[start_idx:start_idx+num_removals]

        # Repair
        for client in removed: 
            best_tour = None
            best_cost = None 

            for idx in range(len(new_tour) + 1): 
                cand_tour = new_tour[:idx] + [client] + new_tour[idx:]
                cand_cost = cost(cand_tour, data)

                if best_cost is None or cand_cost < best_cost:
                    best_tour = cand_tour
                    best_cost = cand_cost

            new_tour = best_tour

        # Accept
        new_cost = cost(new_tour, data)
        if new_cost < curr_cost:
            curr_cost = new_cost
            curr_tour = new_tour

    return curr_tour
```

```python
lns = large_neighborhood_search(data, num_iterations=10000, num_removals=5)
cost(lns, data)
```

```python
draw(df, tour=lns)
```

```python
def large_neighborhood_search_fast(
    data: ProblemData,
    num_iterations: int,
    num_removals: int = 5,
) -> list[int]:
    curr_tour = nearest_neighbor(data)
    curr_cost = cost(curr_tour, data)

    for _ in range(num_iterations):
        cand_tour = copy(curr_tour)
        removed = []

        start = random.randint(0, len(cand_tour) - num_removals)
        removed = cand_tour[start : start + num_removals]
        del cand_tour[start : start + num_removals]

        for client in removed:  # re-insert removed clients
            best_idx, best_cost = None, None

            for idx in range(len(cand_tour) + 1):
                pred = cand_tour[idx - 1] if idx > 0 else cand_tour[-1]
                succ = (
                    cand_tour[idx] if idx != len(cand_tour) else cand_tour[0]
                )
                new_cost = (
                    data.distances[pred, client]
                    + data.distances[client, succ]
                    - data.distances[pred, succ]
                )

                if best_idx is None or new_cost < best_cost:
                    best_idx = idx
                    best_cost = new_cost

            cand_tour.insert(best_idx, client)

        cand_cost = cost(cand_tour, data)
        if cand_cost < curr_cost:
            curr_cost = cand_cost
            curr_tour = copy(cand_tour)

    return curr_tour
```

```python
lns_fast = large_neighborhood_search_fast(data, num_iterations=10000, num_removals=5)
cost(lns_fast, data)
```

```python
draw(df, tour=lns_fast)
```
